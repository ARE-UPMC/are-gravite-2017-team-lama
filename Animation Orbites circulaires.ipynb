```python
%matplotlib notebook
from pylab import *
import numpy as np

from matplotlib import animation



k = 2*pi
w = 0


xmin = -3
xmax = 3
nbx = 100

x = linspace(xmin, xmax, nbx)


xlim(xmin, xmax)
ylim(-3,3)
plot(0, 0, 'yo')
x = cos(k+w)
y = sin(k+w)
x2 = 2*cos(k+w)
y2 = 2*sin(k+w)

for i in range(50):
    plot(x, y, 'wo')
    plot(x2, y2, 'wo')
    w=w+2*pi/50
    x = cos(k+w)
    y = sin(k+w)
    x2 = 2*cos(k+w)
    y2 = 2*sin(k+w)
    line.set_ydata(y)
    line.set_ydata(x)
    plot(x, y, 'bo')
    plot(x2, y2, 'ro')
    draw()
    
show()

       

    
 
```

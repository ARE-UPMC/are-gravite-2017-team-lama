```python
%matplotlib notebook
from pylab import *
import numpy as np

from matplotlib import animation



k = 2*pi
w = 0


xmin = -5
xmax = 5
nbx = 100

x = linspace(xmin, xmax, nbx)


xlim(xmin, xmax)
ylim(-5,5)
plot(0, 0, 'yo')
x = -cos(k+w)
y = -sin(k+w)
x2 = 2*cos(k+w)
y2 = 2*sin(k+w)
x3 = -3*cos(k+w)
y3 = -3*sin(k+w)
x4 = 4*cos(k+w)
y4 = 4*sin(k+w)


for i in range(50):
    plot(x, y, 'wo')
    plot(x2, y2, 'wo')
    plot(x3, y3, 'wo')
    plot(x4, y4, 'wo')
    w=w+2*pi/50
    x = -cos(k+w)
    y = -sin(k+w)
    x2 = 2*cos(k+w)
    y2 = 2*sin(k+w)
    x3 = -3*cos(k+w)
    y3 = -3*sin(k+w)
    x4 = 4*cos(k+w)
    y4 = 4*sin(k+w)
    
    plot(x, y, 'bo')
    plot(x2, y2, 'ro')
    plot(x3, y3, 'bo')
    plot(x4, y4, 'ro')
    draw()
    
show()
```

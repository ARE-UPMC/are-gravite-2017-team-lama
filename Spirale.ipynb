```python
#Affichage d'une spirale

%matplotlib notebook
from pylab import *
import numpy as np

from matplotlib import animation



k = 2*pi
w = 0
t = 1
t2 =-1

xmin = -10
xmax = 10
nbx = 100

x = linspace(xmin, xmax, nbx)


xlim(xmin, xmax)
ylim(-10,10)
plot(0, 0, 'yo')
x = t*cos(k+w)
y = t*sin(k+w)
x2 = t2*cos(k+w)
y2 = t2*sin(k+w)


for i in range(100):
    plot(x, y, 'wo')
    plot(x2, y2, 'wo')
    w=w+2*pi/50
    t=t+1/10
    t2=t2-1/10
    x = t*cos(k+w)
    y = t*sin(k+w)
    x2 = t2*cos(k+w)
    y2 =t2*sin(k+w)
    plot(x, y, 'yo')
    plot(x2, y2, 'ro')
    draw()
    
show()
```
